```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('For loop #')
plt.ylabel('Time (sec)')
plt.title('Parallel matrix multiplication running times (static schedule)')
plt.plot([1, 2, 3], [1.022704, 1.06088525, 1.83623975])
plt.plot([1, 2, 3], [1.022704, 1.06088525, 1.83623975], 'bo')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('For loop #')
plt.ylabel('Time (sec)')
plt.title('Parallel matrix multiplication running times (dynamic schedule)')
plt.plot([1, 2, 3], [1.04917275, 1.10998075, 36.72462375])
plt.plot([1, 2, 3], [1.04917275, 1.10998075, 36.72462375], 'bo')
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot([1, 2, 3], [3.59781625, 3.59781625, 3.59781625])
plt.plot([1, 2, 3], [1.022704, 1.06088525, 1.83623975])
plt.plot([1, 2, 3], [1.04917275, 1.10998075, 36.72462375])


plt.xlabel('For loop #')
plt.ylabel('Time (sec)')
plt.title('Matrix multiplication running times')
plt.legend(("Serial", "Parallel (static schedule)", "Parallel (dynamic schedule)"), loc=0)

plt.show()
```
