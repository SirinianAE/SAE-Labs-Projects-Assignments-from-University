```python
#Collecting Our Data
businesses_in_Torronto = []

with open('business.json') as f:
    for line in f:
        try:
            if(json.loads(line.strip())["city"] == "Toronto" and json.loads(line.strip())["neighborhood"] != ""):
                businesses_in_Torronto.append({x: json.loads(line.strip())[x] for x in ["business_id", "neighborhood"]})
#                businesses_in_Torronto.append(json.loads(line.strip()))
        except ValueError:
            pass
df_businesses_in_Torronto = pd.DataFrame.from_records(businesses_in_Torronto)
```
